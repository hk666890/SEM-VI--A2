```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def exp_programmatic_morphology_simple():
    # STEP 1: Create a binary image with 'Bird' text and noise
    img = np.zeros((100, 400), dtype=np.uint8)
    font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(img, 'Bird', (10, 70), font, 2, 255, 5, cv2.LINE_AA)
    noise = np.random.random((100, 400)) > 0.98
    img[noise] = 255

    # STEP 2: Create a 5x5 square structuring element (kernel)
    kernel = np.ones((5, 5), np.uint8)

    # STEP 3: Apply Morphological Operations
    erosion = cv2.erode(img, kernel, iterations=1)
    dilation = cv2.dilate(img, kernel, iterations=1)
    opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)
    closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)

    # STEP 4: Display Results
    images = [img, erosion, dilation, opening, closing]
    titles = ["Original Programmatic", "Erosion", "Dilation", "Opening", "Closing"]

    plt.figure(figsize=(15, 5))
    for i in range(len(images)):
        plt.subplot(1, 5, i + 1)
        plt.imshow(images[i], cmap='gray')
        plt.title(titles[i])
        plt.axis('off')
    plt.tight_layout()
    plt.show()

# Execute the experiment
exp_programmatic_morphology_simple()
```
